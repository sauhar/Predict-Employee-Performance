```python
import sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
## test_size=0.2 means 20% of data will be used for test and other for training set
```

```python
x.columns
```

```python
#scaled x train
x_train = scaler.fit_transform(x_train)
```

```python
import joblib
#Joblib can be used to parallelize operations, speed up calculations, and reduce overall execution time
joblib.dump(scaler,"scaler.pkl")
#The joblib.dump(scaler, "scaler.pkl") function call saves the scaler object to a file named scaler.pkl
```

```python
#scaled x_test
x_test = scaler.transform(x_test)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformace(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
#fit() is used to train a machine learning model on a dataset. It takes in a dataset, typically represented as a 2D array or matrix, and a set of labels, and then fits the model to the data.
```

```python
log_predictions=log_model.predict(x_test)
```

```python
modelperformace(log_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors":[3,5,7,9,11],
              "weights":["uniform","distance"]}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
```

```python
gridkn.fit(x_train, y_train)
```

```python
gridkn.best_params_
```

```python
predictions = gridkn.predict(x_test)
```

```python
modelperformace(predictions)
```

```python
from sklearn.svm import SVC
svm = SVC()
```

```python
param_grid = {
       "C": [0.01,0.1,0.5,1],
       "kernel": ["linear","rbf","poly"]

}
```

```python
gridsvc = GridSearchCV(svm,param_grid, cv=5)
```

```python
gridsvc.fit(x_train, y_train)
```

```python
y_pred = gridsvc.predict(x_test)
```

```python
modelperformace(y_pred)
```
